### Run the DOM From-StatHead Pipeline

This cell loads and executes the `dom_from_stathead.py` processing script.

- Imports the pipeline dynamically from:
  `src/pipelines/dom_from_stathead.py`

- Uses the input dataset:
  `data/scraper/skill_draftees_2000_2025.csv`

- Generates an enriched output file containing DOM / PDOM / RDOM metrics:
  `data/processed/players_dom_metrics_filled.csv`

- Prints a confirmation message once the pipeline finishes.


```python
from pathlib import Path
import importlib.util
import sys

ROOT = Path("/Users/chasesiegel/Desktop/Comp_Sci/Capstone/Dynasty")

SCRIPT_PATH  = ROOT / "src" / "pipelines" / "dom_from_stathead.py"
PLAYERS_CSV  = ROOT / "data" / "scraper" / "skill_draftees_2000_2025.csv"
OUT_CSV      = ROOT / "data" / "processed" / "players_dom_metrics_filled.csv"

spec = importlib.util.spec_from_file_location("dom_from_stathead", SCRIPT_PATH)
dom_mod = importlib.util.module_from_spec(spec)
sys.modules["dom_from_stathead"] = dom_mod
spec.loader.exec_module(dom_mod)

dom_mod.main(PLAYERS_CSV, OUT_CSV)

print(f"✅ Done. Output saved to: {OUT_CSV}")
```
